```python
import os
import json
import nltk
from rank_bm25 import BM25Okapi
from nltk.tokenize import word_tokenize

# Ensure NLTK tokenizer and other resources are available
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab') # Download the missing 'punkt_tab' resource

# Paths to dataset folders
finished_folder_path = '/content/samples/Finished'
diagnostic_flowchart_folder_path = '/content/diagnostic_kg/Diagnosis_flowchart'


# Check if the paths exist
finished_exists = os.path.exists(finished_folder_path)
diagnostic_flowchart_exists = os.path.exists(diagnostic_flowchart_folder_path)

# Load preprocessed documents
def load_documents(folder_path):
    """Load preprocessed documents from a folder."""
    documents = []
    file_names = []

    for root, _, files in os.walk(folder_path):
        for file in files:
            if file.endswith('.json'):
                file_path = os.path.join(root, file)
                try:
                    with open(file_path, 'r', encoding='utf-8') as f:
                        data = json.load(f)
                        # Assuming the preprocessing step already performed text extraction
                        text = json.dumps(data)  # Convert JSON content to string
                        documents.append(text)
                        file_names.append(file)  # Store file name for reference
                except Exception as e:
                    print(f"Error loading {file_path}: {e}")
    return documents, file_names

# Load documents from both folders
finished_docs, finished_files = load_documents(finished_folder_path)
diagnosis_docs, diagnosis_files = load_documents(diagnostic_flowchart_folder_path)

# Combine both datasets
documents = finished_docs + diagnosis_docs
file_names = finished_files + diagnosis_files

# Verify preprocessing
print(f"Preprocessing completed for all {len(finished_docs)} files in 'Finished' folder.")
print(f"Preprocessing completed for all {len(diagnosis_docs)} files in 'Diagnosis_flowchart' folder.")
print("Preprocessing verification completed.")

# Tokenization for BM25 indexing
tokenized_docs = [word_tokenize(doc.lower()) for doc in documents]

# Initialize BM25
bm25 = BM25Okapi(tokenized_docs)

# Define query expansion function (example: medical term synonyms)
medical_terms_dict = {
    "hyperthyroidism": ["overactive thyroid", "thyroid storm", "graves' disease"],
    "palpitations": ["heart racing", "irregular heartbeat", "arrhythmia"],
    # Add more terms as needed...
}

def expand_query(query):
    """Expand query with medical synonyms."""
    tokens = word_tokenize(query.lower())
    expanded_query = []

    for token in tokens:
        if token in medical_terms_dict:
            expanded_query.extend(medical_terms_dict[token])  # Add synonyms for medical terms
        else:
            expanded_query.append(token)

    return expanded_query

def search_bm25(query, top_n=5):
    """Retrieve top N documents for a given query."""
    expanded_query = expand_query(query)
    tokenized_query = word_tokenize(" ".join(expanded_query).lower())
    scores = bm25.get_scores(tokenized_query)
    top_indices = sorted(range(len(scores)), key=lambda i: scores[i], reverse=True)[:top_n]

    results = [(file_names[i], documents[i][:500], scores[i]) for i in top_indices]
    return results

# Example Query
test_query = "heart palpitations and hyperthyroidism"
results = search_bm25(test_query)

# Display Results
for idx, (file, text, score) in enumerate(results):
    print(f"Rank {idx+1}: {file} (Score: {score})\n{text}\n")
```

```python
import torch
from transformers import AutoTokenizer, AutoModel
import numpy as np
from tqdm import tqdm
import os

# Paths for saving the model and embeddings
embedding_file_path = '/content/doc_embeddings.npy'

# Initialize BioBERT model and tokenizer
model_name = "dmis-lab/biobert-base-cased-v1.1"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

def generate_embeddings(texts):
    """Generate embeddings for the documents using BioBERT."""
    embeddings = []
    for text in tqdm(texts, desc="Generating Embeddings", unit="document"):
        inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=512)
        with torch.no_grad():
            outputs = model(**inputs)
        embedding = outputs.last_hidden_state.mean(dim=1)  # Mean pooling to get sentence embeddings
        embeddings.append(embedding.numpy())

    embeddings = np.array(embeddings)
    return embeddings

# Load documents (from your existing loading function)
documents = finished_docs + diagnosis_docs  # Combine documents from both sources

# Generate embeddings for all documents
doc_embeddings = generate_embeddings(documents)

# Save the embeddings to disk to avoid retraining in the future
np.save(embedding_file_path, doc_embeddings)  # Save embeddings
print("Embeddings saved successfully to disk.")
```

```python
import numpy as np
import faiss

# Path to saved embeddings
embedding_file_path = '/content/doc_embeddings.npy'

# Check if embeddings file exists to avoid retraining
if os.path.exists(embedding_file_path):
    doc_embeddings = np.load(embedding_file_path)
    print("Embeddings loaded from disk.")
else:
    print("Embeddings file not found. Please train the model first.")

# Reshape doc_embeddings to 2D before adding to FAISS index
doc_embeddings = doc_embeddings.reshape(doc_embeddings.shape[0], -1) # Reshape to (num_embeddings, embedding_dim)


# Step 3: FAISS Indexing for Dense Retrieval
dim = doc_embeddings.shape[1]  # Dimension of the embeddings (should match BioBERT output)
index = faiss.IndexFlatL2(dim)  # L2 distance metric for FAISS

# Normalize the document embeddings for cosine similarity
faiss.normalize_L2(doc_embeddings)  # Normalize vectors

# Add document embeddings to FAISS index
index.add(doc_embeddings)  # Add embeddings to the index
print("FAISS index created and embeddings added.")
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# Load Falcon-RW-1B model and tokenizer
model_name = "tiiuae/falcon-rw-1b"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name).to("cpu")  # Use CPU

```

```python
def generate_answer_with_falcon(prompt):
    """Generates a response using Falcon-RW-1B on CPU without repetition."""
    inputs = tokenizer(prompt, return_tensors="pt", truncation=True, max_length=512).to("cpu")

    # Generation settings with stop and anti-repetition
    outputs = model.generate(
        **inputs,
        max_new_tokens=200,
        temperature=0.7,
        top_p=0.9,
        do_sample=True,
        repetition_penalty=1.2,
        pad_token_id=tokenizer.eos_token_id,
        eos_token_id=tokenizer.eos_token_id
    )

    response = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Clean up repeated prompt (if any)
    if prompt in response:
        response = response.split(prompt, 1)[-1].strip()

    return response

# Example
query = "What are the symptoms of thyroid disease?"
print(generate_answer_with_falcon(query))
```
